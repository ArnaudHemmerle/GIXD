<a id='Functions'></a>
### 1.1 Functions

```python
import numpy as np
pi = np.pi
import pandas as pd
from IPython.display import display
from IPython.display import Image
import matplotlib.pyplot as plt
from matplotlib import cm
plt.rcParams["figure.figsize"] = (5,5)

def f_qxy11(a, b, gamma):
    return 2*pi*np.sqrt(a**2+b**2-2*a*b*np.cos(gamma))/(a*b*np.sin(gamma))

def f_qxy1bar1(a, b, gamma):
    return 2*pi*np.sqrt(a**2+b**2+2*a*b*np.cos(gamma))/(a*b*np.sin(gamma))
    
def f_qxy02(a, b, gamma):
    return 4*pi/(b*np.sin(gamma))

def f_qz11(a , b, gamma, psi, theta):
    return -2*pi*np.tan(theta)/(a*b*np.sin(gamma))*(b*np.sin(gamma)*np.cos(psi)-(b*np.cos(gamma)-a)*np.sin(psi))

def f_qz1bar1(a , b, gamma, psi, theta):
    return -2*pi*np.tan(theta)/(a*b*np.sin(gamma))*(b*np.sin(gamma)*np.cos(psi)-(b*np.cos(gamma)+a)*np.sin(psi))

def f_qz02(a , b, gamma, psi, theta):
    return -4*pi/(b*np.sin(gamma))*np.tan(theta)*np.sin(psi)


def gaussian_2D(x, y, x0, y0, A=1, sigmax=.5, sigmay=.5):
    return A*np.exp(-( (x-x0)**2/(2*sigmax**2) + (y-y0)**2/(2*sigmay**2) ))

def peak_2D(x, z, x0, z0, Ax = 1., Az = 1., Lx = 2.5, Lz = 5.):
    
    #Lorentz peak in qxy
    peakx = Ax/(1+np.power((x-x0)*Lx, 2.0))
    
    #Sinc in qz
    temp = (z-z0)*Lz/2.
    peakz = Az*np.power(np.sinc(temp/pi), 2.0)
    
    
    return peakx*peakz

def display_qtable(a , b, gamma, psi, theta, index):
    """Display a table with the q values."""
    
    qxy11 = f_qxy11(a, b, gamma)
    qxy1bar1 = f_qxy1bar1(a, b, gamma)
    qxy02 = f_qxy02(a, b, gamma)
    
    qz11 = f_qz11(a , b, gamma, psi, theta)
    qz1bar1 = f_qz1bar1(a , b, gamma, psi, theta)
    qz02 = f_qz02(a , b, gamma, psi, theta)
    
    
    q_list = [[qxy11, qxy1bar1, qxy02,
               qz11, qz1bar1, qz02]]
    q_df = pd.DataFrame(q_list, columns = ["$q_\parallel^{[11]}$",
                                           "$q_\parallel^{[1\\bar{1}]}$",
                                           "$q_\parallel^{[02]}$",
                                           "$q_z^{[11]}$",
                                           "$q_z^{[1\\bar{1}]}$",
                                           "$q_z^{[02]}$"],
                                index=[index])
 
    q_df = q_df.applymap("{0:.3f}".format)
    
    display(q_df)
    
def plot_qmap(a , b, gamma, psi, theta, index):
    """Plot the map qz vs qparallel, I(qz), and I(qxy)."""

    
    qxy11 = f_qxy11(a, b, gamma)
    qxybar1bar1 = qxy11
    qxy1bar1 = f_qxy1bar1(a, b, gamma)
    qxybar11 = qxy1bar1
    qxy02 = f_qxy02(a, b, gamma)
    qxy0bar2 = qxy02
    
    qz11 = f_qz11(a , b, gamma, psi, theta)
    qzbar1bar1 = -qz11
    qz1bar1 = f_qz1bar1(a , b, gamma, psi, theta)
    qzbar11 = -qz1bar1
    qz02 = f_qz02(a , b, gamma, psi, theta)
    qz0bar2 = -qz02
    
    x = np.linspace(10, 20, 100)
    z = np.linspace(0, 10, 100)
    X,Z = np.meshgrid(x, z)

    I11 = peak_2D(X, Z, qxy11, qz11)
    Ibar1bar1 = peak_2D(X, Z, qxybar1bar1, qzbar1bar1)
    I1bar1 = peak_2D(X, Z, qxy1bar1, qz1bar1)
    Ibar11 = peak_2D(X, Z, qxybar11, qzbar11)
    I02 = peak_2D(X, Z, qxy02, qz02)
    I0bar2 = peak_2D(X, Z, qxy0bar2, qz0bar2)

    I = I11+Ibar1bar1+I1bar1+Ibar11+I02+I0bar2

    #map qz vs qparallel
    fig, ax = plt.subplots()
    p = ax.pcolor(X, Z, I, vmin=0, vmax = np.max(I)*1.01, cmap=cm.afmhot)

    plt.xlabel("$q_{xy}$", fontsize=18)
    plt.ylabel("$q_z$", fontsize=18)
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    plt.title(index)

    #cbar = fig.colorbar(p)
    #cbar.ax.tick_params(labelsize=13) 
    
    plt.show()
    plt.close()
    
    #I(qparallel)
    fig, ax = plt.subplots()
    
    qz_integr = np.sum(I, axis = 0)

    plt.xlabel("$q_{xy}$", fontsize=18)
    plt.ylabel("Sum over $q_z$", fontsize=18)
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    plt.title(index)
    
    plt.plot(x, qz_integr)
    plt.show()

    plt.close()
    
    #I(qz)
    fig, ax = plt.subplots()
    
    qxy_integr = np.sum(I, axis = 1)

    plt.xlabel("$q_z$", fontsize=18)
    plt.ylabel("Sum over $q_{xy}$", fontsize=18)
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    plt.title(index)
    
    plt.plot(x, qxy_integr)
    plt.show()

    plt.close()
    
def plot_lattice(a , b, gamma, psi, theta):

    
    beta = pi-gamma
    ah = 1./2*np.sqrt((b**2+a**2-2*b*a*np.cos(beta)))
    bh = np.sqrt((a**2+b**2-2*a*b*np.cos(gamma))/4)
    omega = np.arccos((ah**2+b**2-bh**2)/(2*ah*b))
    tau = np.arccos((-b**2+a**2+4*bh**2)/(4*a*bh))


    y1 = np.sin(omega)*ah
    x1 = np.cos(omega)*ah

    x_dots = [j*x1 + i*b for j in range(-20,20) for i in range(-20,20)]
    y_dots = [y1*j for j in range(-20,20) for i in range(-20,20)]



    fig, ax = plt.subplots()
    
    plt.plot(y_dots, x_dots,'bo', markersize = 10)
    plt.xlim(-3*b,3*b)
    plt.ylim(-3*b,3*b)

    plt.plot([0,a*np.sin(gamma)],[0,a*np.cos(gamma)], 'c', label = '$a$', lw = 2)

    plt.plot([0,0],[0,b], 'k', label = '$b$', lw = 2)

    if ~np.isclose(theta,0):
        #plt.plot([0+2*a,a*np.cos(psi)+2*a],[0,a*np.sin(psi)], 'r', label = '$n$', lw = 5)
        plt.plot([4*y1,4*y1+a*np.cos(psi)],[4*y1*np.cos(gamma),a*np.sin(psi)+(4*y1)*np.cos(gamma)], 'r', label = '$n$', lw = 5)

        

    plt.legend(fontsize=12, loc='upper left')


    plt.xlabel("$x (nm)$", fontsize=18)
    plt.ylabel("$y (nm)$", fontsize=18)
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)

    plt.title('Direct lattice', fontsize = 15)
    plt.legend(fontsize=12, loc='upper left')

    plt.show()
    plt.close()
    
```

### 1.2 Geometry

The cell is here defined in the most general case of an oblique cell with molecules pointing in a direction out of the plane. 

- $\gamma$ is the angle of distortion.
- $\theta$ is the tilt angle of the molecule, away from the normal vector, with $\theta=0$ for straight molecules.
- $\psi$ is the in-plane inclination of the molecules, with $\psi=0$ aligned along a.
- $a$ and $b$ are the direct vectors.

Molecules are considered as cylindrical, and are invariant by rotation around their principal axes $\bf{n}$.

The unit cell is defined with two molecules, implying that peak ordering follows the one used with a rectangular lattice.

Only the first order peaks $\bf{q^{[11]}}$, $\bf{q^{[1\bar{1}]}}$, and $\bf{q^{[02]}}$ will be considered here.

We use the fact that:
$$q_\parallel^{[hk]}=q_\parallel^{[\bar{h}\bar{k}]}$$
$$q_z^{[hk]}=-q_z^{[\bar{h}\bar{k}]}$$

## 2. Collection of examples

<a id='Hexagonal'></a>
### 2.1 Hexagonal lattice

- $\gamma=\pi/2$
- $b = a \sqrt 3$
- One molecule has 6 NN.

<a id='HexStraight'></a>
#### 2.1.1 Straight molecules

- $\theta=0$
- $\psi=0$ ($\psi$ is not relevant here)

$$q_\parallel^{[11]}=q_\parallel^{[1\bar{1}]}=q_\parallel^{[02]}=\frac{4 \pi}{a \sqrt 3}$$
$$q_z^{[11]}=q_z^{[1\bar{1}]}=q_z^{[02]}=0$$ 

```python
#Hexagonal lattice
#Straight molecules

index = "Hexagonal Straight"

a = 0.5
b = a*np.sqrt(3)
gamma = pi/2
theta = 0
psi = 0

display_qtable(a , b, gamma, psi, theta, index)

plot_qmap(a , b, gamma, psi, theta, index)

plot_lattice(a , b, gamma, psi, theta)  
```

<a id='HexNN'></a>
#### 2.1.2 Molecules inclined towards NN

- $\theta\neq0$
- $\psi=[0; \pi; \pm \pi/3;  \pm 2\pi/3]$

$$q_\parallel^{[11]}=q_\parallel^{[1\bar{1}]}=q_\parallel^{[02]}=\frac{4 \pi}{a \sqrt 3}$$
$$q_z^{[11]}=q_z^{[1\bar{1}]}=\frac{2\pi}{a}\tan{\theta}$$
$$q_z^{[02]}=0$$

```python
#Hexagonal lattice
#Molecules inclined towards NN

index = "Hexagonal NN"

a = 0.5
b = a*np.sqrt(3)
gamma = pi/2
theta = -20*pi/180.
psi = 0


display_qtable(a , b, gamma, psi, theta, index)

plot_qmap(a , b, gamma, psi, theta, index)

plot_lattice(a , b, gamma, psi, theta)  
```

<a id='HexNNN'></a>
#### 2.1.3 Molecules inclined towards NNN

- $\theta\neq0$
- $\psi=[\pm\pi/2]$

$$q_\parallel^{[11]}=q_\parallel^{[1\bar{1}]}=q_\parallel^{[02]}=\frac{4 \pi}{a \sqrt 3}$$
$$2 q_z^{[11]}=q_z^{[02]}$$
$$q_z^{[11]}=q_z^{[1\bar{1}]}=-q_z^{[\bar{1}1]}=\frac{2\pi}{a\sqrt 3}\tan{\theta}$$

```python
#Hexagonal lattice
#Molecules inclined towards NNN

index = "Hexagonal NNN"

a = .5
b = a*np.sqrt(3)
gamma = pi/2
theta = -20*pi/180.
psi = pi/2

display_qtable(a , b, gamma, psi, theta, index)

plot_qmap(a , b, gamma, psi, theta, index)

plot_lattice(a , b, gamma, psi, theta) 
```

<a id='HexInterm'></a>
#### 2.1.4 Molecules inclined towards an intermediate direction

- $\theta\neq0$
- $\psi\neq0$

$$q_\parallel^{[11]}=q_\parallel^{[1\bar{1}]}=q_\parallel^{[02]}=\frac{4 \pi}{a \sqrt 3}$$
$$q_z^{[11]}=\frac{2 \pi \tan{\theta}}{a \sqrt 3} (\sin{\psi} + \sqrt 3 \cos{\psi}) $$
$$q_z^{[\bar{1}1]}=\frac{2 \pi \tan{\theta}}{a \sqrt 3} (\sin{\psi} - \sqrt 3 \cos{\psi})  $$
$$q_z^{[02]}=\frac{4\pi}{a \sqrt 3} \tan{\theta} \sin{\psi}$$

```python
#Hexagonal lattice
#Molecules inclined towards an intermediate direction

index = "Hexagonal Intermediate"

a = .5
b = a*np.sqrt(3)
gamma = pi/2
theta = -30*pi/180.
psi = 40*pi/180.

display_qtable(a , b, gamma, psi, theta, index)

plot_qmap(a , b, gamma, psi, theta, index)

plot_lattice(a , b, gamma, psi, theta) 
```

<a id='Oblique'></a>
### 2.4 Oblique lattice

Most general case.

<a id='ObliqueStraight'></a>
#### 2.4.1 Straight molecules

- $\theta=0$
- $\psi=0$ ($\psi$ is not relevant here)

$$q_\parallel^{[11]}=\frac{2 \pi}{a b \sin{\gamma}}\sqrt{a^2+b^2-2ab\cos{\gamma}}$$
$$q_\parallel^{[1\bar{1}]}=\frac{2 \pi}{a b \sin{\gamma}}\sqrt{a^2+b^2+2ab\cos{\gamma}}$$
$$q_\parallel^{[02]}=\frac{4 \pi}{b \sin{\gamma}}$$

$$q_z^{[11]}=q_z^{[1\bar{1}]}=q_z^{[02]}=0$$ 

```python
#Oblique lattice
#Straight molecules

index = "Oblique Straight"

a = .6
b = 1.
gamma = 70*pi/180
theta = 0
psi = 0


display_qtable(a , b, gamma, psi, theta, index)

plot_qmap(a , b, gamma, psi, theta, index)

plot_lattice(a , b, gamma, psi, theta) 
```

<a id='ObliqueInterm'></a>
#### 2.4.2 Molecules inclined towards an intermediate direction

- $\theta\neq0$
- $\psi\neq0$

$$q_\parallel^{[11]}=\frac{2 \pi}{a b \sin{\gamma}}\sqrt{a^2+b^2-2ab\cos{\gamma}}$$
$$q_\parallel^{[1\bar{1}]}=\frac{2 \pi}{a b \sin{\gamma}}\sqrt{a^2+b^2+2ab\cos{\gamma}}$$
$$q_\parallel^{[02]}=\frac{4 \pi}{b \sin{\gamma}}$$


$$q_z^{[11]}=\frac{2 \pi \tan{\theta}}{a b \sin{\gamma}} (a \sin{\psi} \sin{\gamma} - (a \cos{\gamma} - b) \cos{\psi}) $$
$$q_z^{[1\bar{1}]}=\frac{2 \pi \tan{\theta}}{a b \sin{\gamma}} (-a \sin{\psi} \sin{\gamma} + (a \cos{\gamma} + b) \cos{\psi}) $$
$$q_z^{[02]}=\frac{4\pi}{b} \tan{\theta} \left (\sin{\psi} - \frac{\cos{\psi}}{\tan{\gamma}}\right)$$

```python
#Oblique lattice
#Molecules inclined towards an intermediate direction

index = "Oblique Intermediate"

a = .5
b = .7
gamma = 80*pi/180
theta = -20*pi/180.
psi = 70*pi/180.

display_qtable(a , b, gamma, psi, theta, index)

plot_qmap(a , b, gamma, psi, theta, index)

plot_lattice(a , b, gamma, psi, theta) 
```
